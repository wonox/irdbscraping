```python
# PythonでWeb上の画像などのファイルをダウンロード
# import os
# import pprint
# import time
import urllib.error
import urllib.request

def download_file(url, dst_path):
    try:
        with urllib.request.urlopen(url) as web_file:
            data = web_file.read()
            with open(dst_path, mode='wb') as local_file:
                local_file.write(data)
    except urllib.error.URLError as e:
        print(e)

url = 'https://www.jsps.go.jp/j-grantsinaid/27_kdata/data/3-4-1/3-4-1_r2.pdf'
# windowsではopenの引数にフォルダを指定するとPermissionErrorになる
dst_path = './data/test.pdf'
download_file(url, dst_path)
```

```python
import pandas as pd
import camelot.io as camelot
#pdfファイルを読み込む
table = camelot.read_pdf("./data/test.pdf", pages='all', lattice=True, split_text=True, strip_text='\n')
lists = []  # 一時保存用リスト
for t in table:
  lists.append(t.df)
  df = pd.concat(lists)  # データフレームを結合

df
# 42.4s
```

```python
df2 = df
df2.columns = ["機関名","採択件数（件）","女性比率（％）","40歳未満比率（％）","直接経費（千円）","間接経費（千円）","合計（千円）","基盤研究Ａ","基盤研究B","基盤研究C","新規応募件数（件）","新規採択件数（件）","新規採択率（％）"]
# df[df['機関名']=="機関名"]
# df[df['機関名'].isnull()]
# print(df2[df2['機関名'] == ''])
# print(df2[df2['機関名'] == '機関名'])
# print(df2[(df2['機関名'] == '')|(df2['機関名'] == '機関名')])
# df2['機関名'].isnull()
df_and = df2[(df2['機関名'] == '機関名') | (df2['機関名'] == '')]
# df_and.index
df3 = df2.drop(df_and.index)
df3.reset_index(inplace=True, drop=True)
# ok df3['機関名'] = df3['機関名'].str.replace('＊', '')
df3.loc['機関名':,:] = df3.loc['機関名':,:].replace('＊', '')
df3
```
